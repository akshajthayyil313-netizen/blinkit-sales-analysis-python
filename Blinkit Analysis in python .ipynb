### **BUSINESS REQUIREMENTS**

### **KPI'S REQUIREMENTS**

```python
#Total Sales
total_sales = df['Sales'].sum()

##Average Sales
avg_sales = df['Sales'].mean()

# No of Item Sold
no_of_item_sold = df['Sales'].count()

#Average Ratings
avg_ratings = df['Rating'].mean()

#Display

print(f"Total Sales: ${total_sales:,.0f}")
print(f"AverageSales: ${avg_sales:,.0f}")
print(f"No of Items Sold: {no_of_item_sold:,.0f}")
print(f"Average Ratings: {avg_ratings:,.0f}")
```

### **CHARTS REQUIREMENT**

### **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat,labels=sales_by_fat.index,
                       autopct='%1.1f%%',
                       startangle= 90)

plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

### **Total Sales by Item Type**

```python
import matplotlib.pyplot as plt

# 1. Calculate the sales data
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# 2. Set up the figure size (This must be on its own line)
plt.figure(figsize=(10, 6))

# 3. Create the bar chart (This must be on its own line)
bars = plt.bar(sales_by_type.index, sales_by_type.values)

# 4. Apply labels and formatting
plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

# 5. Add the data numbers on top of each bar
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, 
             bar.get_height(), 
             f'{bar.get_height():,.0f}', 
             ha='center', 
             va='bottom')

# 6. Display the final plot
plt.tight_layout()
plt.show()
```

### **Fat Content by Outlet for Total Sales**

```python
# 1. Group the data and unstack it (Split this from the next command)
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

# 2. Reorder columns (This needs its own line)
# Note: Ensure these names match your data exactly (e.g., 'Low Fat' or 'LF')
grouped = grouped[['Regular', 'Low Fat']] 

# 3. Create the plot (Split this from the label command)
ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content')

# 4. Add labels (These need their own lines)
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')

# 5. Add legend and show plot
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

### **Total Sales by Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

# 1. Set figure size (Split from plot command)
plt.figure(figsize=(9, 5))

# 2. Create the line plot
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment')

# 3. Add text labels (Loop is fine on one line, but better split for readability)
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom')

# 4. Show plot (Split these two commands!)
plt.tight_layout()
plt.show()
```

### **Sales by Outlet Size**

```python
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4, 4)) # Removed the stray 'I'

# Split the pie chart creation from the title
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%1.1f%%', startangle=90)
plt.title('Sales by Outlet Size') # New line

plt.show()
```

### **Sales by Outlet Location**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Split the calculation and the sorting (Press Enter!)
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

# 2. Setup the figure
plt.figure(figsize=(8, 3))

# 3. Create the plot
# IMPORTANT: This must be on a new line. Do not put it after a '#' symbol.
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

# 4. Show the plot (Split these two commands!)
plt.tight_layout()
plt.show()
```
